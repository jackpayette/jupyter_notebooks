```python
import pickle as pkl
from matplotlib import pyplot as plt
import os
import numpy as np

%cd /work/eggNOG/
```

```python
condensed_matrices = []
groups = []
for filename in os.listdir('chloroflexi_cyano_dists-simple_25th_percentile//'):
    if not filename.endswith('.pkl'):
        continue
    condensed_matrices.append(pkl.load(open('chloroflexi_cyano_dists-simple_25th_percentile/%s' % filename, 'rb')))
    groups.append(filename.replace('.pkl', ''))
```

```python
quartile1 = []
quartile3 = []
medians   = []
for _ in condensed_matrices:
    medians.append(np.median(_))
    quartile1.append(np.percentile(_, 25))
    quartile3.append(np.percentile(_, 75))
```

```python
fig, axs = plt.subplots(figsize=(50, 20), nrows=4, sharey=False, dpi=300)

lower_thresh = np.mean(quartile1) - np.std(quartile1)
upper_thresh = np.mean(quartile3) + np.std(quartile3)

for ax, position in zip(axs, range(0, len(groups), 200)):
    ax.fill_between(range(0, 202), y1=upper_thresh, y2=lower_thresh, alpha=0.1, color='r')
    ax.boxplot(condensed_matrices[position:position+200],
               showfliers=False);
    ax.set_xticklabels(groups[position:position+200], rotation=45, ha='right');
fig.tight_layout()
```
