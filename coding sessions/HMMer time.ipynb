```python
import pandas as pd
import subprocess
import ete3
import numpy as np
import re
```

```python
header       = 'target_name accession query_name accession seq_E-value seq_score seq_bias domain_E-value domain_score domain_bias exp reg clu ov env dom rep inc'.split()
descriptions = []
df_data      = []
for line in open('/Users/thiberio/RedA.tab_yeah').readlines():
    if line.startswith('#'):
        continue
    line   = line.strip()
    values = line.split()[:18]
    df_data.append(values)
    descriptions.append(' '.join(line.split()[18:]))
df = pd.DataFrame(columns=header, data=df_data)
```
