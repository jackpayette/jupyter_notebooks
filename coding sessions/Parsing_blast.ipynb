```python
header = ['query acc.ver', 'subject acc.ver', '% identity',
          'alignment length', 'mismatches', 'gap opens', 
          'q. start', 'q. end', 's. start', 's. end', 'evalue',
          'bit score', 'subject tax id']
```

```python
df = pd.read_table('Proteo_4540.bls', sep='\t', comment='#',
                   header=None, names=header)
df.info()
```

```python
df['coverage %'] =  (df['q. end'] - df['q. start'] + 1) / 485.
df = df[((df['% identity'] >= 80)|
        (df['coverage %'] >= 0.98))&
       (df['gap opens'] <= 2)]
```

```python
for query_name in df['query acc.ver'].unique()[:1]:
    print(query_name)
    tmp_df = df[df['query acc.ver'] == query_name].copy()
    out = open('yeah.hit_ids', 'w')
    out.write('\n'.join(tmp_df['subject acc.ver'].unique()))
    out.close()
    os.system('blastdbcmd -db blast/nr/nr -entry_batch yeah.hit_ids > recruited_hits')
    break
```
