```python
import re
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
whole_alignment_well_supported_branches = []
count = 0
for support, br_length in re.findall(
    '\)([^(),:]+):(\d+?\.\d+?)[,()]',
    open('/work/site_rate/hug_et_al/whole_alignment.treefile').read()):
    alrt, bb = support.split('/')
    if float(alrt) >= 80 and float(bb) >= 95:
        whole_alignment_well_supported_branches.append(float(br_length))
```

```python
no_slow_evolving_well_supported_branches = []
count = 0
for support, br_length in re.findall(
    '\)([^(),:]+):(\d+?\.\d+?)[,()]',
    open('/work/site_rate/hug_et_al/selected_categories/bestTree/no_slow_sites.aln-LG.treefile').read()):
    alrt, bb = support.split('/')
    if float(alrt) >= 80 and float(bb) >= 95:
        no_slow_evolving_well_supported_branches.append(float(br_length))
```

```python
fig, ax = plt.subplots(dpi=300, figsize=(10,5))
sns.kdeplot(whole_alignment_well_supported_branches, label='whole alignment', bw=0.001, shade=True, ax=ax)
sns.kdeplot(no_slow_evolving_well_supported_branches, label='no slow evolving sites', bw=0.001, shade=True, ax=ax)
```
