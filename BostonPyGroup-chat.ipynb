```python
from scipy.stats import hypergeom #hypergeometric test
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import random
import seaborn as sns
from scipy.stats import mannwhitneyu
```

```python
yeah = [random.sample(range(100), 10) for n in range(10)]
```

```python
yeah
```

```python
df = pd.DataFrame(index=['V%i' %i for i in range(10)], columns=['S%i' %i for i in range(10)], data=yeah)
df
```

```python
fig, ax = plt.subplots(figsize=(7,4), dpi=300)
for index, row in df.iterrows():
    sns.kdeplot(row.values, shade=True, label=index)
```

```python
mannwhitneyu(df.loc['V2'].values, df.loc['V5'].values)
```
